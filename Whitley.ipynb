```python
import random
import decimal
import struct
import math
random.seed(3)
```

```python
def generate_pop(n,k,m):
    pop = []
    i = 0
    while (i < n):
        pk = []
        j = 0
        while(j < k):
            p = random.uniform(m[j][0],m[j][1])
            j+=1
            pk.append(p)
        pop.append(pk)
        i+=1
    return pop

    
```

```python
def select(pr,pop):
    p1 = random.randint(0,len(pop)-1)
    p2 = random.randint(0,len(pop)-1)
    pop.sort(key=f)
    w = random.randint(int(pr*len(pop)),len(pop)-1)
    pop[w] = crossover(pop[p1],pop[p2])
    return pop 
```

```python
def mutate(p,m,sh):
    for i in range(len(p)):
        z = random.uniform(0,1)/sh
        pr = random.random()
        if(pr>0.5):
            p[i]+=z
            if(p[i]>m[i][1]):
                p[i] = m[i][1]
        else:
            p[i]-=z
            if(p[i]<m[i][0]):
                p[i] = m[i][0]
    return p 
```

```python
def crossover(p1,p2):
    i = 0
    child = []
    for i in range(len(p1)):
        pr = random.random()
        if(pr < 0.5):
            child.append(p1[i])
        else:
            child.append(p2[i])       
    return child      
```
