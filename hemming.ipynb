```python

def msgToBin(msg):
    bin_msg = ''
    for x in msg:
        tmp=format(ord(x), 'b')
        if len(tmp)==6:
            tmp='0'+tmp
        a = ''.join(tmp)
        bin_msg = bin_msg + a
    global leng
    leng = len(bin_msg)
    return bin_msg

        
def binToMsg(b=None):
    bb = []
    while len(b) > 7:
        bb.append(b[:7])
        b = b[7:]
    bb.append(b)
    return ''.join([chr(int(x, 2)) for x in bb])


def codeBlock(block,k):
    block = list(map(int,block))
    block.insert(0,0)
    x = 1
    i = 1
    while i < k:
        block.insert(x,0)
        i+=1
        x = x*2 + 1
    i = 0
    while i < k:
        s = 0
        step = 2**i
        j = step - 1
        while j < len(block):
            for z in range(step):
                if(j+z)>=len(block):
                    break
                s  = s+ block[j + z]
            j = j + 2*step
        block[step-1] = s%2
        i += 1
    s = 0
    for ch in block:
        s+=ch
#     print("summa ",s)
#     print("last bit",s%2)
    block.append(s%2)
    return block

def codeMsg(s,m,k):
    sToBin = msgToBin(s)
    while(len(sToBin)%m != 0):
        sToBin = sToBin+'0'
    codeS = ''
    for i in range(len(sToBin)//m):
        st = ''.join(str(e) for e in codeBlock(sToBin[m*i:m*i+m],k))
        codeS = codeS+st
    return codeS


def decodeBlock(block,m,k):
    block = list(map(int,block))
    lastB = block[-1]
#     print("last bit",lastB)
    block = block[:-1]
    summ = 0
    for e in block:
        if e == 1:
            summ+=1
#     print("kontr summa ",summ)
    summ = summ%2
    newBlock = copy.copy(block)
    newBlock = list(map(int,newBlock))
    newBlock[0] = 0
    x = 1
    i = 1
    while i < k:
        newBlock[x]=0
        i+=1
        x = x*2 + 1
    i = 0
    while i < k:
        s = 0
        step = 2**i
        j = step - 1
        while j < len(block):
            for z in range(step):
                if(j+z)>=len(block):
                    break
                s  = s + newBlock[j + z]
            j = j + 2*step
        newBlock[step-1] = s%2
        i += 1
    num = 0
    for x in range(len(block)):
        if(block[x]!=newBlock[x]):
            num +=x+1
    num -=1
    if(num==-1 and summ ==lastB):
        m = 1
      #  print("no error")
    elif(num!=-1 and summ!=lastB):
        #print("error in ",num," bit")
        if(num<len(block)):
            if(newBlock[num]==1):
                newBlock[num]=0
            else:
                newBlock[num]=1
    elif(num!=-1 and summ ==lastB):
        m = 2
       # print("double error:(")
    newBlock[0] = '9'
    x = 1
    i = 1
    while i < k:
        newBlock[x]='9'
        i+=1
        x = x*2 + 1
    stBlock = ''.join(map(str,newBlock)).replace('9','')
    return stBlock
def decodeMsg(code,m,k):
    decode = ''
    for i in range(len(code)//(m+k+1)):
        decode = decode + decodeBlock(code[(m+k+1)*i:(m+k+1)*(i+1)],m,k)
    msg = binToMsg(decode[0:leng])
#     msg=msg.replace('_',' ')
    print("ispravlennoe:")
    print(msg)
    return msg

def doMistake(num,code,m,k):
    code = list(code)
    badS = ''
    for i in range(len(code)//(m+k+1)):
        for j in range(num):
            x = random.randint((m+k+1)*i,(m+k+1)*(i+1)-1)
#             print("mistake were made in ",x%(m+k+1),"bit")
            if(code[x]=='1'):
                code[x]='0'
            else:
                code[x]='1'
#         print("start: ",(m+k+1)*i, " end ",(m+k+1)*(i+1)-1)
        s = code[(m+k+1)*i:(m+k+1)*(i+1)]
#         print(len(s))
        s = s[:-1]
        s[0] = '9'
        x = 1
        o = 1
        while o < k:
            s[x]='9'
            o+=1
            x = x*2 + 1
        s = ''.join(map(str,s)).replace('9','')
        badS = badS + s
    print("with mistake:")
    print(binToMsg(badS[0:leng]))
    code = ''.join(code)
    
    return code
    
    
    
    
    
    
```
